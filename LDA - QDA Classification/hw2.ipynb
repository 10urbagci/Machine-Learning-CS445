```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
import pandas as pd
import numpy as np
df = pd.read_csv("gdrive/My Drive/Colab Notebooks/CS445/penguins.txt",delimiter = "\t")
```

```python
#independent features
X = df[['bill_length_mm','flipper_length_mm']]
#Dependent features
y = df['species']
```

```python
#Data traning
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
#Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
#Confusion Matrix for LDA
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
acc = accuracy_score(y_test,y_pred)
print('Confusion Matrix for LDA')
print(cm)
print(acc)
```

```python
#Quadratic Discrimant Analaysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
clf = QuadraticDiscriminantAnalysis()
clf.fit(X_train, y_train)
```

```python
qda_pred = clf.predict(X_test)
```

```python
qda_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,qda_pred)
acc = accuracy_score(y_test,qda_pred)
print('Confusion Matrix for QDA:')
print(cm)
print(acc)
```

```python
#LDA Covariance Matrix
lda = LinearDiscriminantAnalysis(store_covariance=True).fit(X_train,y_train)
print('Linear:')
print(lda.covariance_)
```

```python
#QDA Covarience Matrix
qda = QuadraticDiscriminantAnalysis(store_covariance=True).fit(X_train,y_train)
print('Quadratic:')
print(qda.covariance_)
```
